```python
import csv
from sklearn.decomposition import PCA
from kmodes.kmodes import KModes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
import json
import os
import statsmodels.api as sm
import statsmodels.stats.api as sms
from math import sqrt, pi, sin, cos, asin

final_dataset = pd.read_csv("inc_dataset.csv",index_col=0)

print(final_dataset.shape)
#print(final_dataset)

#PCA!
name_to_fam = dict()
name_to_loc = dict()
locs = list()
#import families
with open('languoid.txt') as id_to_fam:
    csv_reader = csv.reader(id_to_fam,delimiter = '\t')
    next(csv_reader)
    for row in csv_reader:
        #print(row)
        lat = float(row[3])
        long = float(row[4])
        name_to_loc[row[0]]=np.array([lat,long])
        name_to_fam[row[0]]=row[-5]
lats=list()
longs = list()


#for our MSE calculations
for name in final_dataset.index:
    lats.append(name_to_loc[name][0])
    longs.append(name_to_loc[name][1])
    
    
def a2r(angle):
    return angle / 180 * pi

#I added great cirlce distance after making  the charts so the numbers are different
#but there is no real differece. Geography doesn't map onto the PCA graph well.
def great_circle_dist(lon0, lat0, lon1, lat1):
    out = []
    earth_radius_m = 6378137

    lon_diff = lon1 - lon0
    lat_diff = lat1 - lat0
    a = (sin(a2r(lat_diff / 2))**2
                 + cos(a2r(lat1)) * cos(a2r(lat0)) * sin(a2r(lon_diff / 2))**2)
    c = 2 * asin(sqrt(a))
    d = c * earth_radius_m
    return d
def geo_dist(lang1,lang2):
    loc1 = name_to_loc[lang1]
    loc2 = name_to_loc[lang2]
    return great_circle_dist(loc1[1],loc1[0],loc2[1],loc2[0])


#preprocessing, NA vals -> mean
def pca(final_dataset):
    imp_mean=SimpleImputer(missing_values=np.nan, strategy='mean')
    dataset=imp_mean.fit_transform(final_dataset)
    pca_graph(dataset)

#dictionary from datset to the language family
fam_ls = list()
for name in final_dataset.index.values:
    try:
        fam = name_to_fam[name]
        fam_ls.append(name_to_fam[name])
    except:
        fam_ls.append("other")

        
#adding a column for language family
fam_col = pd.DataFrame({'Category':fam_ls})

fam_col.index=list(final_dataset.index.values)

df = final_dataset.join(fam_col)

fam_col = pd.DataFrame({'Category':fam_ls})
print(fam_col)
print(df)

#calculates pca graph, histogram for mse and mse
def pca_graph(dataset):
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(dataset)
    overall_diff=None
    max_m=0
    print(principalComponents.T.shape)
    print(np.asarray(lats).shape)
    #Linear regression lines for lats and longs
    model = sm.OLS(np.asarray(lats),np.asarray(principalComponents)).fit()
    print(model.summary())
    model = sm.OLS(np.asarray(longs),np.asarray(principalComponents)).fit()
    print(model.summary())
    best_ds = list()
    
    #try many different values for m to see which scale minimizes the mse of the PCA graph
    for m in range(0,10000000,1000):
        tot_diff = 0
        count=0
        ds = list()
        for i in range(len(final_dataset.index.values)):
            for j in range(len(final_dataset.index.values)):
                lang1 = final_dataset.index.values[i]
                lang2 = final_dataset.index.values[j]
                pca1 = np.array(principalComponents[i])
                pca2 = np.array(principalComponents[j])
                pca_dist = np.linalg.norm(pca1-pca2)
                g_dist = geo_dist(lang1,lang2)
                diff = (g_dist - (m)*pca_dist)**2
                tot_diff+=diff
                if g_dist !=0:
                    count+=1
                    ds.append(diff)
        tot_diff = tot_diff/count
        #print(tot_diff)
        
        #print(np.var(ds))
        if overall_diff == None or tot_diff<overall_diff:
            overall_diff=tot_diff
            max_m=m
            best_ds = ds
    print(overall_diff)
    print(max_m)
    _ = plt.hist(ds)
    plt.show()
    print(sms.DescrStatsW(best_ds).tconfint_mean())
    
    principalDf = pd.DataFrame(data = principalComponents, columns = ['c1', 'c2'])
    principalDf = principalDf.join(fam_col)
    plot = sns.pairplot(x_vars=['c1'],y_vars=['c2'],hue='Category',data=principalDf)
    plt.show()


for filename in os.listdir(os.getcwd()+'/var_ds/'):
    print(filename)
    pca(pd.read_csv('./var_ds/'+filename,index_col=0))
```
